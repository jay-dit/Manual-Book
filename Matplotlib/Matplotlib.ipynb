```python
import numpy as np
import matplotlib.pyplot as plt

#  :
x = np.linspace(0, 10, 100)
y = np.sin(x)
img = y.reshape(5, 20)

#   "Figure"  "Axes":
fig, axes = plt.subplots(nrows = 2, ncols = 1)

#  ,  :
axes[0].plot(x, y)
axes[1].imshow(img)

#   :
axes[0].set(title = 'sin(x)')
axes[1].set(title = 'img')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3)

for ax in axes:
    mu = 10*np.random.random()
    kappa = 10*np.random.random()
    data = np.random.vonmises(mu, kappa, size=1000)
    ax.hist(data, bins = 50, rwidth = 0.4)

#      :
axes[0].set_title('loc = "left", pad = 0',
                loc = 'left',
                pad = 0)    #     

axes[1].set_title('loc = "center", pad = 10',
                loc = 'center',     #    
                pad = 10)

axes[2].set_title('loc = "right", pad = 20',
                loc = 'right',
                pad = 20)
fig.suptitle('',
                fontsize = 20,
                y = 1.2)
fig.set_figheight(2)
fig.set_figwidth(14)

plt.show()

#  fontsize   ,   color  backgroundcolor       :
#           fontfamily  fontstyle:
#     Axes         Figure,       suptitle():
```

```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 1.1, 2.9
data = np.random.normal(mu, sigma, 10000)

fig, ax = plt.subplots()

ax.hist(data, bins = 50, rwidth = 0.4)
ax.text(x = 4, y = 500, s = r'$\frac{1}{\sqrt{2\pi\sigma^{2}}}e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}$',
        fontsize = 20)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3*np.pi, 3*np.pi, 200)
y1 = np.sin(x) - 2
y2 = np.cos(x) + 2
y3 = np.sinc(x)

fig, ax = plt.subplots()

ax.plot(x, y1, label = 'sin(x)')
ax.plot(x, y2, label = 'cos(x)')
ax.plot(x, y3, label = r'$\frac{sin(x)}{x}$')

ax.legend()

fig.set_figheight(5)
fig.set_figwidth(8)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3*np.pi, 3*np.pi, 200)
y1 = np.sin(x)
fig, axes = plt.subplots(5, 2)

location = ['upper right', 'upper left', 'lower left',
            'lower right', 'right', 'center left', 
            'center right', 'lower center', 'upper center', 'center']
i = 0

for ax in axes.ravel():
    ax.plot(x, y1, label = 'sin(x)')
    ax.legend(loc = location[i])
    ax.set_title(location[i])
    ax.set_xticks([])
    ax.set_yticks([])
    i += 1

fig.set_figheight(10)
fig.set_figwidth(10)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 10)

y1 = 2*x + 5
y2 = -x + 8
y3 = 0.5*x + 5
y4 = np.full(10, 5)

fig, ax = plt.subplots()

ax.plot(x, y1, label = 'y = 2x + 5')
ax.plot(x, y2, label = 'y = -x + 8')
ax.plot(x, y3, label = 'y = 0.5x + 5')
ax.plot(x, y4, label = 'y = 5')

ax.legend(fontsize = 25,
          ncol = 2,    #   
          facecolor = 'oldlace',    #   
          edgecolor = 'r',    #    
          title = '',    #  
          title_fontsize = '20'    #    
         )

fig.set_figwidth(12)
fig.set_figheight(12)

plt.show()
```

```python
#axis
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3*np.pi, 3*np.pi, 200)
y = np.sinc(x)

fig = plt.figure()

ax_1 = fig.add_subplot(2, 1, 1)
ax_2 = fig.add_subplot(2, 2, 3)
ax_3 = fig.add_subplot(2, 2, 4)

ax_1.plot(x, y)
ax_1.grid(axis = 'x')
ax_1.set_title('axis = "x"')

ax_2.plot(x, y)
ax_2.grid(axis = 'y')
ax_2.set_title('axis = "y"')

ax_3.plot(x, y)
ax_3.grid(axis = 'both')
ax_3.set_title('axis = "both"')

fig.set_figwidth(12)
fig.set_figheight(12)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3*np.pi, 3*np.pi, 200)
y = np.sinc(x)

fig, ax = plt.subplots()

ax.plot(x, y, linewidth = 5)

ax.grid(color = 'blue',    #   
        linewidth = 2,    #  
        linestyle = '--')    #  

fig.set_figwidth(12)
fig.set_figheight(6)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3*np.pi, 3*np.pi, 200)
y = np.sinc(x)

fig, ax = plt.subplots()

ax.plot(x, y,
        color = 'r',
        linewidth = 4)

#      
#     
#  :
ax.minorticks_on()

#       :
ax.grid(which='major',
        color = 'k', 
        linewidth = 2)

#      
#  :
ax.grid(which='minor', 
        color = 'k', 
        linestyle = ':')

fig.set_figwidth(12)
fig.set_figheight(6)

plt.show()
```
