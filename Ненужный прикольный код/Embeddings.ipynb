```python
df = {el:navec[el] for el in navec.vocab.words}
```

```python
def top_N(n, N = 10):
    global df
    dicti = {key:sum((val - n)**2) for key, val in df.items()}
    return sorted(dicti.items(), key=lambda item: item[1])[:N]
```

```python
top_N(navec['еда'] - navec['вкус'], 20)
```

```python
top_N(navec['доктор'] - navec['мужчина'] + navec['женщина'], 50)
```
