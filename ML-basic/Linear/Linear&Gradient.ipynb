## Gradient descent

```python
import numpy as np 
y = lambda x: np.cos(x)
dy = lambda x: -np.sin(x)
Nu = 0.1
n1 = -0.5
c = []
for j in range(200):
    n1 = n1 - Nu*dy(n1)
    c.append(n1)
c1 = np.array(c)
c2 = y(c1)
a = np.arange(-4, 2, 0.01)
b = y(a)
import matplotlib.pyplot as plt
plt.plot(a,b, label = 'sin(x)')
plt.scatter(c1, c2, c = 'orange', label = 'LR')
plt.legend(loc = 'lower right')
plt.show()
print('\nТочка минимума: ',n1,'\n\nМинимум:',y(n1))
```

## Linear regression

```python
aX = np.arange(0, 10)
y = lambda x: 2*x + 3 
aY = y(aX) + np.random.choice([-1,1])*np.random.sample(10)*2
```

```python

k = 1
b = 1
Nu = 0.0001

for i in range(100000):
    summ = 0
    for key, val in enumerate(aY):
        summ += 2*(key*k+b-val)*key
    k = k - Nu*summ
    
    summ = 0
    for key, val in enumerate(aY):
        summ += 2*(key*k+b-val)
    b = b - Nu*summ
    # print(k,b)
y2 = lambda x: k*x + b
aY2 = y2(aX)
import matplotlib.pyplot as plt
plt.scatter(aX, aY, c = 'orange', label = 'LR')
plt.plot(aX, aY2)
plt.show()
print(k,b)
```

## Gini

```python
income = {'poor': 0.1429, "middle class": 0.2857, "rich": 0.5714}
cum_income = {(i+1)/len(income) : list(income.values())[i]/sum(list(income.values()))  for i in range(len(income))}
S = 0
for k, v in cum_income.items():
    S += v
    cum_income[k] = S
XY = {0: 0}
XY.update(cum_income)
print(XY)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots()
axes.plot(XY.keys(), XY.values())
axes.plot([0, 1, 1, 0],[0, 1, 0, 0])
plt.show()
```
