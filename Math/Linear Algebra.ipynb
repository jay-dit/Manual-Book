### Условия существования обратной матрицы:
    
    1) Матрица должна быть квадратной 
    2) Столбцы линейно независимы

Система векторов называется линейно зависимой, если из этих векторов можно составить нулевую (равную нулю) линейную комбинацию, т. е., причем хотя бы один из коэффициентов линейной комбинации отличен от нуля.

Линейная комбинация - сумма векторов, где каждый умножен на скаляр. Получается результирующий вектор.

Линейная оболочка - множество всех линейных комбинаций
